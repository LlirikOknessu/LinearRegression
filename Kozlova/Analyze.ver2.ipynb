```python
# Grouped Bar Chart
plt.figure(figsize=(7,9))
# set width of bar
barWidth = 0.25
# set height of bar
y1= df.Life_expectancy.where(df.Year == 2002).dropna()
y2 = df.Life_expectancy.where(df.Year == 2005).dropna()
# Set position of bar on X axis
pos1 = np.arange(len(y1))
pos2 = [x + barWidth for x in pos1]
# Make the plot
plt.bar(pos1, y1, color='#FBC02D', width=barWidth, label='Batting')
plt.bar(pos2, y2, color='#F57F17', width=barWidth, label='Bowling')
# Add xticks on the middle of the group bars
plt.xlabel('Assignee', fontweight='bold')
plt.ylabel('Number of Tickets', fontweight='bold')


# Create legend & Show graphic
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(30,10))
barWidth = 0.25
settings = df['Country'].unique()
avr_pre = []
for setting in settings:
    avr_pre.append(df['Life_expectancy_'].loc[df['Country'] == setting].mean())
bars =plt.bar(settings, avr_pre, color='#1168A8', label='Country')
plt.xticks(rotation=90)
for bar, value in zip(settings, avr_pre):
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2, height + 5, value, ha="center", va="bottom"
    )
plt.legend()
plt.show()
```
