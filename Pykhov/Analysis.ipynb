```python
plt.figure(figsize=(15,10))

corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix,2)
corr_matrix[np.abs(corr_matrix)<0.3] = 0

sb.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation Matrix')
plt.show()
```
