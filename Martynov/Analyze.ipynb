```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/raw/Summary_of_Weather.csv')
df.head()
```

```python
df = df[['Date', 'Precip', 'MaxTemp', 'MinTemp', 'MeanTemp', 'Snowfall', 'YR', 'MO']]
df.head()
```

```python
corr = df.corr()
sns.heatmap(corr, square=True)
```

```python
df.info()
```

```python
sns.regplot(x='MinTemp', y = 'MaxTemp', data = df)
sns.set_style('whitegrid')
plt.ylim(0,)
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.distplot(df['MaxTemp']);
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.distplot(df['MinTemp']);
```

  Анализ данных показал, что коррелируют между собой лишь несколько столбцов данных - напрямую связанные с температурой и временем, а также с осадками в виде снега и чистыми осадками в миллиметрах. Были построены распределения по темературам для определения медиальных и модальных значений.
