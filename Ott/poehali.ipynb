```python
from sklearn.impute import SimpleImputer
```

```python
cols_with_nan = [col for col in df.columns
                 if df[col].isnull().any()]

reduced_df = df.drop(cols_with_nan, axis=1)
cols_with_nan
```

```python
df = df.drop(columns='Country')
```

```python
#developing - 0, developed - 1
df['Status'] = df['Status'].apply(lambda x: 0 if x == 'Developing' else 1)
```

```python
df.describe()
```

```python
df_plus = df.copy()
for col in cols_with_nan:
    df_plus[col+ '_was_missing'] = df_plus[col].isnull()

df_plus.info()
```

```python
my_imputer = SimpleImputer()

df_imputed = pd.DataFrame(my_imputer.fit_transform(df_plus))

df_imputed.columns = df_plus.columns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error as mae
from sklearn.model_selection import train_test_split as tts
```

```python
X = df_imputed.drop('Life expectancy ', axis=1)
```

```python
y = df_imputed['Life expectancy ']
```

```python
X_train, X_test, y_train, y_test = tts(X, y, train_size=0.8,
                                               random_state=42)
```

```python
model = LinearRegression()
```

```python
history = model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
```

```python
err = mae(y_test, preds)
print (err)
```

```python
model2 = Ridge()
```

```python
model2.fit(X_train, y_train)
print(mae(y_test, model2.predict(X_test)))
```

```python
y_baseline = np.random.uniform(y_test.min(), y_test.max(), len(y_test))
y_baseline
```

```python
mae(y_test, y_baseline)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_tree = DecisionTreeRegressor()
model_tree.fit(X_train, y_train)
preds = model_tree.predict(X_test)
err = mae(y_test, preds)
print (err)
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
param_grid = {
    'max_depth': np.arange(2, 100, 8, dtype = int),
    'splitter': ['best', 'random'],
    'min_samples_split': np.arange(2, 50, 10),
    'min_samples_leaf': np.arange(2, 50, 10)
}

search = GridSearchCV(model_tree, param_grid, n_jobs=-1, refit = True)
search.fit(X_train, y_train)
print(search.best_params_)
```

```python
preds = search.best_estimator_.predict(X_test)
err = mae(y_test, preds)
print (err)
```
